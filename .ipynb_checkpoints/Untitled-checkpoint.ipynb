```python
import numpy as np
import pandas as pd 
import os
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
#balancing dataset

dataNoStrok = df[df.stroke != 1]
dataYesStroke = df[df.stroke != 0]
dataNoStrok = dataNoStrok.sample(300)
dataset = pd.concat([dataNoStrok,dataYesStroke],axis=0)
dataset.shape
```

```python
plt.figure(figsize=[15,17])
fft=["gender","age","hypertension","heart_disease","ever_married","work_type","Residence_type","avg_glucose_level","bmi","smoking_status"]
n=1
for f in fft:
    plt.subplot(6,2,n)
    sns.histplot(x=f, hue='stroke', edgecolor="black", alpha=0.7, multiple="stack", data=dataset)
    sns.despine()
    plt.title("Countplot of Strokes  by {}".format(f))
    n=n+1
plt.tight_layout()
plt.show()
```

```python
sm = dataset.smoking_status[dataset.stroke == 1].value_counts()
sns.barplot(x = sm.index, y = sm.values, alpha=0.7)
```

```python
def gender_conv(x):
    if x == 'Male' : return 1
    if x == 'Female' : return 2
    return 0

def marriage_conv(x):
    if x == 'Yes' : return 1
    if x == 'No' : return 2
    return 0

def work_type_conv(x):
    if x == 'Private'        : return 1
    if x == 'Self-employed'  : return 2
    if x == 'Govt_job'       : return 3
    if x == 'children'       : return 4
    return 0

def residence_type(x):
    if x == 'Urban' : return 1
    if x == 'Rural' : return 2
    return 0

def smokingStatus_conv(x):
    if x == "never smoked " : return 1
    if x == "formerly smoked" : return 2
    if x == "smokes" : return 3
    return 0
```

```python
df.gender         =  df.gender.apply(gender_conv)
df.ever_married   =  df.ever_married.apply(marriage_conv)
df.work_type      =  df.work_type.apply(work_type_conv)
df.Residence_type =  df.Residence_type.apply(residence_type)
df.smoking_status =  df.smoking_status.apply(smokingStatus_conv)
df.head(100)
```

```python
gender1 = {'Male':1,'Female':2, 'Other':0}
df.gender = [gender1[item] for item in df.gender]

marriage1 = {'Yes':1,'No':2}
df.ever_married = [marriage1[item] for item in df.ever_married]

work1 = {'children':1, 'Govt_job':2, 'Never_worked':3, 'Private':4,'Self-employed':5}
df.work_type = [work1[item] for item in df.work_type]

residence1 = {'Urban':1,'Rural':2}
df.Residence_type = [residence1[item] for item in df.Residence_type]

smoking1 = {'never smoked': 1, 'formerly smoked':2, 'smokes':3, 'Unknown':0}
df.smoking_status = [smoking1[item] for item in df.smoking_status]
```

```python
print("Any missing values in training set:",dataset.isnull().values.any())
```

# Treat Missing Values with Mean Replacement

```python
df.gender              = df.gender.fillna(dataset.gender.mean())
df.age                 = df.age.fillna(dataset.age.mean())
df.hypertension        = df.hypertension.fillna(dataset.hypertension.mean())
df.heart_disease       = df.heart_disease.fillna(dataset.heart_disease.mean())
df.ever_married        = df.ever_married.fillna(dataset.ever_married.mean())
df.work_type           = df.work_type.fillna(dataset.work_type.mean())
df.Residence_type      = df.Residence_type.fillna(dataset.Residence_type.mean())
df.avg_glucose_level   = df.avg_glucose_level.fillna(dataset.avg_glucose_level.mean())
df.bmi                 = df.bmi.fillna(dataset.bmi.mean())
df.smoking_status      = df.smoking_status.fillna(dataset.smoking_status.mean())
```

```python
features = ["gender","age", "hypertension", "heart_disease", "ever_married", "work_type", "Residence_type", "avg_glucose_level", "bmi",  "smoking_status" ]
X = dataset[features]
Y = dataset.stroke
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=5)
logistic_regression= LogisticRegression()
logistic_regression.fit(X_train,Y_train)
Y_pred=logistic_regression.predict(X_test)
print('Accuracy: ',metrics.accuracy_score(Y_test, Y_pred))
```
